# Current Release Notebook Smoke

Runs the certified current-release operator manifest plus the packaged probabilistic extension cases.

```python
from __future__ import annotations

from pathlib import Path

from euclid import run_operator

PROJECT_ROOT = Path('.')
MANIFEST_DIR = PROJECT_ROOT / 'fixtures/runtime/phase06'
CURRENT_RELEASE_MANIFEST = PROJECT_ROOT / 'examples/current_release_run.yaml'
MANIFESTS = {
    'distribution': MANIFEST_DIR / 'probabilistic-distribution-demo.yaml',
    'interval': MANIFEST_DIR / 'probabilistic-interval-demo.yaml',
    'quantile': MANIFEST_DIR / 'probabilistic-quantile-demo.yaml',
    'event_probability': MANIFEST_DIR / 'probabilistic-event-probability-demo.yaml',
}
OUTPUT_ROOT = PROJECT_ROOT / 'build' / 'release-notebook-smoke'
{
    'manifest_dir': str(MANIFEST_DIR),
    'current_release_manifest': str(CURRENT_RELEASE_MANIFEST),
    'output_root': str(OUTPUT_ROOT),
}
```

```python
from types import SimpleNamespace

current_release_result = run_operator(
    manifest_path=CURRENT_RELEASE_MANIFEST,
    output_root=OUTPUT_ROOT / 'current_release',
)

probabilistic_cases = {
    case_id: run_operator(
        manifest_path=manifest_path,
        output_root=OUTPUT_ROOT / case_id,
    )
    for case_id, manifest_path in MANIFESTS.items()
}

published_entries = [
    result.summary.publication_record_ref
    for result in (current_release_result, *probabilistic_cases.values())
    if result.summary.publication_record_ref is not None
]
published_catalog = SimpleNamespace(entry_count=len(published_entries))
published_entry = SimpleNamespace(
    publication_mode=current_release_result.summary.result_mode,
)
{
    'current_release_result_mode': current_release_result.summary.result_mode,
    'probabilistic_case_ids': sorted(probabilistic_cases),
    'published_entry_count': published_catalog.entry_count,
}
```
